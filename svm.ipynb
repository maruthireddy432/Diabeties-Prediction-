Importing Modules


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data collection and Analysis


```python
diabetes_df=pd.read_csv(r'D:\skill vertex\Major_project\diabetes.csv')
diabetes_df
```

```python
diabetes_df.columns
```

```python
diabetes_df.describe()
```

The dataframe contains 0's in the columns replacing them into null values 

```python
#replacing 0's to null values
diabetes_df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']]=diabetes_df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']].replace(0,np.nan)
```

```python
diabetes_df.isnull().sum()
```

Filling those null values with mean and median

```python
diabetes_df['Glucose'].fillna(diabetes_df['Glucose'].mean(),inplace=True)
diabetes_df['BloodPressure'].fillna(diabetes_df['BloodPressure'].mean(),inplace=True)
diabetes_df['SkinThickness'].fillna(diabetes_df['SkinThickness'].median(),inplace=True)
diabetes_df['Insulin'].fillna(diabetes_df['Insulin'].median(),inplace=True)
diabetes_df['BMI'].fillna(diabetes_df['BMI'].median(),inplace=True)
```

```python
diabetes_df.head(5)
```

```python
diabetes_df['Outcome'].value_counts()
```

```python
diabetes_df.groupby('Outcome').mean()
```

seperating the data and labels 

```python
x=diabetes_df.drop(columns='Outcome',axis=1)
y=diabetes_df['Outcome']
```

Data standardization

```python
scaler=StandardScaler()
standardized_data=scaler.fit_transform(x)
```

```python
standardized_data
```

```python
x=standardized_data
y=diabetes_df['Outcome']
```

train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the model

```python
classifier=svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=200)
rfc.fit(x_train,y_train)
```

Model evaluation
Accuracy score

```python
x_train_prediction=rfc.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy score of training data:',training_data_accuracy)
```

```python
x_test_prediction=rfc.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy score of testing data:',testing_data_accuracy)
```

predicting values

```python
#input data to the model
input_data=(0,131.0,72.40518417462484,29.0,125.0,43.2,0.27,26)
#changing data into numpy array
input_data_numpy_array=np.asarray(input_data)
#reshape the array 
reshaped_array=input_data_numpy_array.reshape(1,-1)
std_data=scaler.transform(reshaped_array)
print(std_data)
prediction=rfc.predict(std_data)
print(prediction)
if prediction[0]==0:
    print('The person non diabetic')
else:
    print('The person is diabetic')
```
